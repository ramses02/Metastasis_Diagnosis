```python
x_train, y_train, x_valid, y_valid, x_test, y_test = None, None, None, None, None, None


for filename, dataset_name in file_to_dataset.items():
    file_path = os.path.join(folder_path, filename)
    
    with h5py.File(file_path, 'r') as file:
        
        dataset = file[dataset_name]
        
        
        if 'train_x.h5' in filename:
            x_train = dataset[:]
        elif 'train_y.h5' in filename:
            y_train = dataset[:]
        elif 'valid_x.h5' in filename:
            x_valid = dataset[:]
        elif 'valid_y.h5' in filename:
            y_valid = dataset[:]
        elif 'test_x.h5' in filename:
            x_test = dataset[:]
        elif 'test_y.h5' in filename:
            y_test = dataset[:]
```

```python
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import numpy as np

# Step 1: Resize Images (using Pillow)
desired_image_size = (64, 64)  # Set your desired image size

x_train_resized = [np.array(Image.fromarray(img).resize(desired_image_size)) for img in x_train]
x_valid_resized = [np.array(Image.fromarray(img).resize(desired_image_size)) for img in x_valid]
x_test_resized = [np.array(Image.fromarray(img).resize(desired_image_size)) for img in x_test]

# Step 2: Normalize Pixel Values
x_train_resized = np.array(x_train_resized) / 255.0
x_valid_resized = np.array(x_valid_resized) / 255.0
x_test_resized = np.array(x_test_resized) / 255.0

# Step 3: Data Augmentation (Optional, using Keras)
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
datagen.fit(x_train_resized)

# Step 4: Splitting the Data
x_train_final, x_val_final, y_train_final, y_val_final = train_test_split(
    x_train_resized, y_train, test_size=0.1, random_state=42
)
```
